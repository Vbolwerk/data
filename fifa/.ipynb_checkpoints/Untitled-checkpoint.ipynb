```python
import csv
import numpy as np
import scipy.stats as stats
import seaborn as sns
with open("fifa_countries_audience.csv", "r") as f:
    reader = csv.reader(f.read().splitlines())
    fifa = []
    for row in reader:
        fifa.append(row)
```

```python
countries = [x[0] for x in fifa[1:]]
confederations = [x[1] for x in fifa[1:]]
population_share = [float(x[2]) for x in fifa[1:]]
tv_audience_share = [float(x[3]) for x in fifa[1:]]
gdp_weighted_share = [float(x[4]) for x in fifa[1:]]
```

```python
dict_fifa={}
for x in range(len(countries)):
    dict_fifa[countries[x]]=confederations[x], population_share[x], tv_audience_share[x], gdp_weighted_share[x]
```

```python
mean_pop=np.mean(population_share)
median_pop=np.median(population_share)
mode_pop= stats.mode(population_share)

mean_audience = np.mean(tv_audience_share)
median_audience = np.median(tv_audience_share)
mode_audience = stats.mode(tv_audience_share)

mean_gdp_share = np.mean(gdp_weighted_share)
median_gdp_share = np.median(gdp_weighted_share)
mode_gdp_share = stats.mode(gdp_weighted_share)

print(mean_pop, median_pop, mode_pop)
print(mean_audience, median_audience, mode_audience)
print(mean_gdp_share, median_gdp_share, mode_gdp_share)
```
